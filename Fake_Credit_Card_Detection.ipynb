Importing necessary dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
data=pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

```python
#Printing first 5 rows
data.head()
```

```python
#information of data
data.info()
```

```python
#distributing legit and fraud transaction
data['Class'].value_counts()
```

```python
#This dataset is not balanced so we need to separate data
legit= data[data.Class==0]
fraud= data[data.Class==1]
```

```python
#we will do statistical measure of the amount in data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#some observation are that value in legits are more than value in fraud
data.groupby('Class').mean()
```

```python
#We will do under sampling as class =0(61327) is very less compared to the data where class =1(163)
legit1=legit.sample(n=492)
```

```python
data1=pd.concat([legit1,fraud],axis=0)
```

```python
#Now we must have same number of legal and fraud sample
data1['Class'].value_counts()
```

```python
data1.groupby('Class').mean()
```

We will be splitting data and features

```python
x=data1.drop(columns='Class',axis=1)
y=data1['Class']
```

Now we will split this data into test data and train data

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

We will mow use logistic regression model for this project

```python
#Training the model with training data
model=LogisticRegression()
model.fit(x_train,y_train)
```

We will evaluate model by the accuracy score

```python
training=model.predict(x_train)
Accuracy=accuracy_score(training,y_train)
print("Accuracy on training data : ",Accuracy)
```

```python
rtrain=model.predict(x_test)
tAccuracy=accuracy_score(rtrain,y_test)
print("Accuracy on test data : ",tAccuracy)
```
